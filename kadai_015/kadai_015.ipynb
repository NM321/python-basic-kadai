```python
class Human:
    def __init__(self):
        self.name = ""
        self.age = ""

    def set(self, name, age):
        self.name = name
        self.age = age

    def printinfo(self):
        print(self.name)
        print(self.age)

Man = Human()
Man.set("Yumi", 18)
Man.printinfo()
```
