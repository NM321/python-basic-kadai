```python
class Human:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def check_adult(self):
        if self.age >= 20:
            print(f"{self.name + "は大人である"}")
        else:
            print(f"{self.name + "は大人ではない"}")

Man1 = Human("Ami", 20)
Man2 = Human("Isana", 15)
Man3 = Human("Umi", 25)

human_data = [Man1, Man2, Man3]

num = 0

for num in range(0, 3):
  human_data[num].check_adult()
```
